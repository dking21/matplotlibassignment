```python
fig, axs = plt.subplots(1,2,figsize=(5,5))

linear = lambda x: 3*x - 0.5
sqrt = lambda x: 5*np.sqrt(x)

functions = [linear,sqrt]
x=np.linspace(0,5,num=50)

for f, ax, z in zip(functions, axs.flatten(),['linear','sqrt']):
    ax.plot(x,f(x))
    ax.legend([z])
    #ax.set_xscale('log')
```

```python
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
fig, ax = plt.subplots()
ax.bar(np.arange(len(barheights)), barheights)
x_pos = np.arange(len(barheights))
ax.set_xticks(x_pos)
ax.set_xticklabels(barlabels, rotation=45)
```
